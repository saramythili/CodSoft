```python
#Plot for Sepal Length
from scipy.stats import gaussian_kde
plt.subplot(2,2,1)
species=dataset['species'].unique()
colors=['pink','red','yellow']
for i, sp in enumerate(species):
    subset=dataset[dataset['species']== sp]['sepal_length']
    kde=gaussian_kde(subset)
    x_vals=np.linspace(min(subset),max(subset), 200)
    plt.plot(x_vals,kde(x_vals), label=sp, color=colors[i])
plt.title("Sepal Length")
plt.legend()
```

```python
#Plot for Sepal Width
from scipy.stats import gaussian_kde
plt.subplot(2,2,2)
for i, sp in enumerate(species):
    subset=dataset[dataset['species']== sp]['sepal_width']
    kde=gaussian_kde(subset)
    x_vals=np.linspace(min(subset),max(subset), 100)
    plt.plot(x_vals,kde(x_vals), label=sp, color=colors[i])
plt.title("Sepal Width")
plt.legend()
```
